```python
buzzword_encoding = pd.DataFrame(columns = range(30))
```

```python
for i in range(len(set(buzzwords_in_papers['index'].tolist()))):
    tmp = buzzwords_in_papers[buzzwords_in_papers['index'] == i]
    
    grouped = tmp.groupby('buzz_id').sum()
    
    row = [0]*30
    
    for j in grouped.index.tolist():
        row[j] += grouped.loc[j]['count']
    buzzword_encoding.loc[len(buzzword_encoding.index)] = row
        
```

```python
buzzword_encoding.to_csv(os.getcwd()+'/data/buzzword_encoding.csv', index = False)
```
