```python
def getPapers(volume):
#     driver = webdriver.Safari()
    
    driver.get(volume)
    return [p.get_attribute('href') for p in driver.find_elements(By.CLASS_NAME,'article-content-title')]
    
```

```python
def scrapeDOI(paper):
#     driver = webdriver.Safari()
    driver.get(paper)
    
    return driver.find_element(By.CLASS_NAME, 'doi').get_attribute('href').split('doi.org/')[1]
    
```

```python
def getTitles(volume):
#     driver = webdriver.Safari()
    
    driver.get(volume)
    return [p.get_attribute('innerHTML') for p in driver.find_elements(By.CLASS_NAME,'js-article-title')]
    
    
    
    
```
