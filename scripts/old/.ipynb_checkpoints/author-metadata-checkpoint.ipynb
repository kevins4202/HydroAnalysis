```python
all_authors = pd.read_csv(os.getcwd().replace('scripts', 'data')+'/author_metadata.csv')
```

```python
no_data = pd.read_csv(os.getcwd().replace('scripts', 'data')+'/missing_authors.csv')['au_id'].tolist()
```

```python
processed = set(pd.read_csv(os.getcwd().replace('scripts', 'data')+'/author_metadata.csv')['id'].tolist())

```

```python
keyi = 0
i = 20857
while i < papers.shape[0]:
    print(i)
    auth_ids = papers.loc[i]['authors'].split(', ')
    
#     print(set(auth_ids))
    
    j = 0
    
    while j < len(set(auth_ids)):
        subt = False
        
        auid = auth_ids[j]
        if auid not in processed:
            key = keys[keyi]
            
            js = requests.get(f"https://api.elsevier.com/content/author/author_id/{auid}?apiKey={key}&httpAccept=application%2Fjson")
            r = js.json()

#             print(r)
#             print("----------------")
            try:
                eid = r['author-retrieval-response'][0]['coredata']['eid']

                citations = r['author-retrieval-response'][0]['coredata']['citation-count']

                citedby = r['author-retrieval-response'][0]['coredata']['cited-by-count']

                names = r['author-retrieval-response'][0]['author-profile']['preferred-name']

                name = names['given-name'] + ' '+ names['surname']

                all_authors.loc[len(all_authors.index)] = [auid, eid, name, citedby, citations]

                processed.add(auid)
            except Exception as e:
                print(str(e))
                no_data.append(auid)
                j+=1
            keyi = (keyi+1)%10
        j+=1
    i+=1
    
    
            
            
            
```
