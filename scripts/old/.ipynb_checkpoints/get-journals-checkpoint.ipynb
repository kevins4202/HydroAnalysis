```python
vols = {}
def getVolumes(journal):
    web_driver = webdriver.Safari()
    
    web_driver.get(journal)
    
    last_url = web_driver.current_url
    times = 0
    
    while True:
        web_driver.implicitly_wait(1)
        web_driver.refresh()
        web_driver.find_elements(By.CLASS_NAME, "accordion-panel-title")[4].click()
        years = [e.get_attribute("innerHTML") for e in web_driver.find_elements(By.CLASS_NAME, "accordion-title")]
#         years = editions[4:]
#         print(len(years))

        volume_buttons = web_driver.find_elements(By.CLASS_NAME, "accordion-panel-title")
#         web_driver.find_elements(By.CLASS_NAME, "accordion-panel-title")[4].click()
        

        for i in range(4, len(volume_buttons)):
            web_driver.find_elements(By.CLASS_NAME, "accordion-panel-title")[i].click()
            indiv_volumes = web_driver.find_elements(By.CLASS_NAME, "js-issue-item-link")
            vols[years[i]] = [indiv_volumes.get_attribute('href') for indiv_volumes in indiv_volumes]
            
#         print(web_driver.find_elements(By.CLASS_NAME, "button-alternative-secondary")[2].get_property('disabled'))
        
        if web_driver.find_elements(By.CLASS_NAME, "button-alternative-secondary")[2].get_property('disabled'):
            break
        else:
#             last_url = web_driver.current_url
#             times+=1
            web_driver.find_elements(By.CLASS_NAME, "button-alternative-secondary")[2].click()
    
    web_driver.close()
    
    return vols
```

```python
getVolumes('https://www.sciencedirect.com/journal/journal-of-hydrology/issues')
```
