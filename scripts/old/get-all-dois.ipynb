```python
import os
import selenium
from selenium import webdriver
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
from selenium.webdriver.common.by import By
```

```python
journals = pd.read_csv(os.getcwd().replace('scripts/old', 'data')+'/journals.csv')
```

```python
journals
```

```python
df = pd.read_csv(os.getcwd().replace('scripts/old', 'data')+'/all_papers.csv')
```

```python
df
```

```python
for i in range(230,journals.shape[0]):
    volumes_in_journal = pd.read_csv(os.getcwd().replace('scripts', 'data')+f'/journals/{i}/volumes.csv')
    papers_in_journal = pd.read_csv(os.getcwd().replace('scripts', 'data')+f'/journals/{i}/papers.csv')
    
    print(papers_in_journal)
    
    driver = webdriver.Safari()
    
    for j in range(papers_in_journal.shape[0]):
        print(f'journal: {i} paper i: {j}')
        #every paper
        link = papers_in_journal.iat[j,1]
        
        print(link)
        
        
        voli = papers_in_journal.iat[j, 0]
        
        if volumes_in_journal.iat[voli,0] < 2018:
            break
        
        driver.get(link)
        
        if('Editorial' in driver.find_element(By.CLASS_NAME, 'title-text').get_attribute('innerHTML')):
            driver.close()
            driver = webdriver.Safari()
            continue
        
        doi = driver.find_element(By.CLASS_NAME, 'doi').get_attribute('href').split('doi.org/')[1]
        
        
        df.loc[len(df.index)] = [i, voli, volumes_in_journal.iat[voli,0], link, doi]
        
    driver.close()
```
