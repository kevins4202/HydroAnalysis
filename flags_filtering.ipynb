```python
red_subj = ['Smart Cities', 'Insurtech', 'Regtech', 'Predictive Analysis', 'Nanotechnology', 'Biotechnology', 'Cybersecurity', 'Data Mining']

green_subj = ['AI and Machine Learning', 'Big Data', 'Data Analytics', 'Web Applications']
```

```python
filtered_papers.columns
```

```python
rf_journals = ['Ecohydrology & Hydrobiology'
'Ecological Modelling'
'Journal of Contaminant Hydrology'
'Journal of Water Process Engineering']
```

```python
def rm(list_r, list_g, rf_journal):
#     print(list_r)
    if rf_journal:
        perc = 0.65
    else:
        perc = 0.5
    print(sum(list_g))
    return (sum(list_g) < perc)
        
```

```python
remove_i = []
```

```python
for i in range(filtered_papers.shape[0]):
    row = filtered_papers.loc[i]
    
    reds = [int(x) for x in row[red_subj]]
    greens = [int(x) for x in row[green_subj]]
    
    sum_r = sum(reds)
    sum_g = sum(greens)
    
#     p_r = round(sum_r/(sum_r+sum_g),2)
#     p_g = round(sum_g/(sum_r+sum_g),2)
    
    percentlist_r = (np.array(reds)/(sum_r+sum_g)).tolist()
    percentlist_g = (np.array(greens)/(sum_r+sum_g)).tolist()
    
    journal = row['journal']
    
    rf_j = bool(journal in rf_journals)
    
    if rm(percentlist_r, percentlist_g, rf_j):
        remove_i.append(row['index'])
```
